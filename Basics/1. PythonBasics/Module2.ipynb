**Question 2**

```python
"""
    Find the number of occurrences of each unique value in `LABEL` (count of values).
    
    Add a column `BINARY` to the DataFrame which contains 1 if `LABEL` is 2,
    else 0.
    
    Find the mean, minimum, maximum and standard deviation of `RESP` for each value in `LABEL`.
"""
```

```python
def calc():
    temp = np.array(dataframe['LABEL'])
    unique = np.unique(temp)
    values = dict()
    for i in unique:
        values[i] = 0
    
    for i in temp:
        if i in unique:
            values[i] = values[i]+1
    
    for i in values:
        print(i, " occurs", values[i],end = '\n')
    print('\n')
    
def binary():
    temp = list()
    for i in dataframe['LABEL']:
        if i == 2.0:
            temp.append(1)
        else:
            temp.append(0)
    dataframe['Binary'] = temp
    
def stats():
    temp = dataframe['LABEL']
    unique = np.unique(temp)
    values = dict()
    for i in unique:
        values[i] = dataframe[(dataframe.LABEL == i)]
    
    for j in values:
        i = values[j]
        print("For the value: ",j)
        print("\tMaximum element = ",i['RESP'].max())
        print("\tMinimum elemet = ",i['RESP'].min())
        print("\tMean = ",(i['RESP'].sum())/len(i))
        i = np.array(i)
        print("\tStandard Deviation = ",np.std(i), end = '\n\n')
    
```

```python
calc()
binary()
stats()
print(dataframe)
```

## Task 2

Some simple file operations.

```python
"""
    Generate a text file containing 50 integer elements between 1 and 20 (use `random`). 
    Identify the unique numbers and the number of their occurrences, and write this into 
    a new text file.
    
    Example
    -------
    
    Generated File
    --------------
    1 2 3 4 3 2 1 ...
    
    Count File
    ----------
    1 - 2
    2 - 2
    3 - 2
    4 - 1
    ...
"""
```

```python
from collections import defaultdict
def rand():
    file = np.random.randint(low = 1, high = 20, size = (50,1))
    fin = open("text.txt",'w')
    for i in file:
        fin.write(str(i[0])+' ')
    fin.close()
    
    fout = open("text.txt",'r')
    count = defaultdict(int)
    data = fout.read()
    data = data.split()
    for j in data:
        count[j] = count[j]+1
    
    fin = open("new.text",'w')
    for j in count:
        fin.write(str(j)+'-'+str(count[j])+'\n')
        print(str(j)+'-'+str(count[j]))
```
