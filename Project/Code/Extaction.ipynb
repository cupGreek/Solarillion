```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as ply
import json
```

```python
airport_codes = ["ATL","CLT","DEN","DFW","EWR","IAH","JFK","LAS","LAX","MCO","MIA","ORD","PHX","SEA","SFO"]
```

```python
cols = ["FlightDate", "Origin", "Dest", "ArrTime", "DepTime", "ArrDel15", "ArrDelayMinutes", "DepDelayMinutes"]

flight_data = pd.DataFrame(columns = cols)
for year in ('2016','2017'):
    for month in range(1,13):
        temp = pd.read_csv("/home/cupgreek/Documents/flight_details/"+year+"/On_Time_On_Time_Performance_"+year+"_"+str(month)+"/On_Time_On_Time_Performance_"+year+"_"+str(month)+".csv", usecols = cols)
        temp = pd.DataFrame(temp)
        temp = temp[temp["Dest"].isin(airport_codes)]
        temp = temp[temp["Origin"].isin(airport_codes)]
        flight_data = pd.concat([flight_data,temp], sort = False, ignore_index = True)
```

```python
flight_data.isnull().values.sum()
```

```python
flight_data = flight_data.dropna()
flight_data.isnull().values.sum()
```

```python
flight_data.sort_values(by='ArrTime', na_position='first')
```

```python
flight_data.sort_values(by='DepTime', na_position='first')
```

```python
flight_data.to_csv('/home/cupgreek/Documents/flight_details(without NaN).csv', index = False)
```

```python
cols = ['windspeedKmph','winddirDegree','weatherCode','precipMM','visibility','pressure','cloudcover','DewPointF','WindGustKmph','tempF','WindChillF','humidity','time']
weather_data = pd.DataFrame(columns = cols)

for airport in airport_codes:
    for year in ('2016','2017'):
        for month in range(1,13):
            with open("/home/cupgreek/Documents/weather/"+airport+"/"+year+"-"+str(month)+".json", "r") as read_file:
                data = json.load(read_file)
                for i in range(len(data['data']['weather'])):
                    date = data['data']['weather'][i]['date']
                    temp_weather = pd.DataFrame(data['data']['weather'][i]['hourly'], columns = cols)
                    temp_weather['FlightDate'] = date
                    temp_weather['airport'] = data['data']['request'][0]['query'][:3]
                    weather_data = pd.concat([weather_data,temp_weather], sort = False, ignore_index = True)
```

```python
weather_data.to_csv('/home/cupgreek/Documents/Solarillion_Project/weather_data.csv', index = False)
```
